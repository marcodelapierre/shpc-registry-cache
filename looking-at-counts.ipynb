```python
import matplotlib.pyplot as plt
import json
```

```python
with open('counts.json', 'r') as fd:
    counts = json.loads(fd.read())
```

```python
plt.figure(figsize=(12,8))
plt.hist(list(counts.values()), bins = 100, log=True, color="darkorange")
plt.title("Frequency (y) of BioContainers executables by count (x)", fontsize=20)
plt.ylabel("Frequency", fontsize=18)
plt.xlabel("Count", fontsize=18)
plt.tight_layout()

fig1 = plt.gcf()
plt.show()
plt.draw()
fig1.savefig('biocontainers-counts.png', dpi=300)
```
